```python
city_to_country = {
    'Madrid': 'Spain',
    'Berlin': 'Germany',
    'New York': 'USA',
    'Toronto': 'Canada',
    'Paris': 'France',
    'Unknown': 'Unknown',
}
city_to_state = {
    'New York': 'NY',
    'Toronto': 'ON',
    'Madrid': 'Undefined',
    'Berlin': 'Undefined',
    'Paris': 'Undefined',
    'Unknown': 'Undefined',
}
country_to_states = {
    'USA': ['NY', 'CA', 'FL', 'TX'], 
    'Canada': ['BC', 'ON'],       
    'Spain': ['Undefined'],
    'Germany': ['Undefined'],
    'France': ['Undefined'],
    'Unknown': ['Undefined']
}
```

```python
def fix_geographic_data(df):
    df_fixed = df.copy()
    
    for index, row in df_fixed.iterrows():
        city = row['City']
        state = row['State']
        
        # Якщо відоме місто, то визначити країну та штат по ньому
        if city != 'Unknown' and city in city_to_country:
            correct_country = city_to_country[city]
            correct_state = city_to_state[city]
            
            df_fixed.at[index, 'Country'] = correct_country
            df_fixed.at[index, 'State'] = correct_state
            
        # Якщо місто не відоме, але відомий штат, визначити країну по ньому
        elif city == 'Unknown' and state != 'Undefined':
            for country, states in country_to_states.items():
                if state in states:
                    df_fixed.at[index, 'Country'] = country
                    break
    
    return df_fixed
```

```python
products_to_categories = {
    'iPhone 14': 'Electronics',
    'Office Chair': 'Furniture',
    'Desk Lamp': 'Home Decor',
    'Gaming Table': 'Furniture',
    'Binder Set': 'Office Supplies'
}
products_to_subcategories = {
    'iPhone 14': 'Phones',
    'Office Chair': 'Chairs',
    'Desk Lamp': 'Lamps',
    'Gaming Table': 'Tables',
    'Binder Set': 'Binders'
}
categories_to_subs = {
    'Books': [],
    'Electronics': ['Phones'],
    'Toys': [],
    'Office Supplies': ['Binders'],
    'Home Decor': ['Lamps'],
    'Furniture': ['Tables', 'Chairs'],
    'Unknown': ['Unknown']
}
```

```python
def fix_product_data(df):
    df_fixed = df.copy()
    
    for index, row in df_fixed.iterrows():
        product = row['ProductName']
        subcategory = row['SubCategory']
        
        # Якщо відомий товар, то визначити категорію та підкатегорію
        if product != 'Unknown' and product in products_to_categories:
            correct_category = products_to_categories[product]
            correct_subcategory = products_to_subcategories[product]
            
            df_fixed.at[index, 'ProductCategory'] = correct_category
            df_fixed.at[index, 'SubCategory'] = correct_subcategory
            
        # Якщо товар не відомий, але відома підкатегорія, визначити категорію
        elif product == 'Unknown' and subcategory != 'Unknown':
            for category, subcategories in categories_to_subs.items():
                if subcategory in subcategories:
                    df_fixed.at[index, 'ProductCategory'] = category
                    break
    
    return df_fixed
```
